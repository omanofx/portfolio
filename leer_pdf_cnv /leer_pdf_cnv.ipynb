```python
pdf_path = "/content/449526.pdf"
```

# Esto puede funcionar

```python
# Extraer las tablas del archivo PDF
tablas = tabula.read_pdf(pdf_path,
                         pages= 0,
                         lattice=True,
                         multiple_tables=True)

# Mostrar las tablas extraídas
for tabla in tablas:
    print(tabla)
```
